# Ques 4  
Simulate and analyze traffic movement on a single road with several lanes using lists.  
- generate random vehicles
- simulate entry and exit.
- measure congestion ,flow and average wait time.

```python
import random
import time

LANES = 3             # number of lanes
ROAD_LENGTH = 20      # each lane length (number of positions)
SIM_TIME = 30         # number of simulation steps
ENTRY_PROB = 0.4      # probability of new vehicle entering a lane
EXIT_POS = ROAD_LENGTH - 1

# Initialize lanes
lanes = [[None for _ in range(ROAD_LENGTH)] for _ in range(LANES)]
vehicle_id = 0
vehicles_data = {}  # {id: {"entry_time": t, "exit_time": t2}}

# Statistics
exited_vehicles = 0
total_wait_time = 0

print(" Traffic Simulation Starting...\n")

for t in range(SIM_TIME):
    print(f"\n--- Time Step {t+1} ---")

    # Move vehicles forward (from end to start)
    for lane in lanes:
        for pos in range(ROAD_LENGTH-1, -1, -1):
            if lane[pos] is not None:
                # Exit condition
                if pos == EXIT_POS:
                    vid = lane[pos]
                    lane[pos] = None
                    vehicles_data[vid]["exit_time"] = t
                    exited_vehicles += 1
                    total_wait_time += vehicles_data[vid]["exit_time"] - vehicles_data[vid]["entry_time"]
                # Move forward if next spot empty
                elif lane[pos+1] is None:
                    lane[pos+1] = lane[pos]
                    lane[pos] = None

    # Randomly add new vehicles
    for i in range(LANES):
        if lanes[i][0] is None and random.random() < ENTRY_PROB:
            vehicle_id += 1
            lanes[i][0] = vehicle_id
            vehicles_data[vehicle_id] = {"entry_time": t}
    
    # Print the road state
    for i, lane in enumerate(lanes):
        lane_view = ''.join(['' if cell else '路' for cell in lane])
        print(f"Lane {i+1}: {lane_view}")
    
    time.sleep(0.2)  # slow down simulation for readability

# --- Results ---
total_vehicles = len(vehicles_data)
congestion = sum(sum(1 for x in lane if x) for lane in lanes) / (LANES * ROAD_LENGTH)
avg_wait_time = total_wait_time / exited_vehicles if exited_vehicles > 0 else 0
flow = exited_vehicles / SIM_TIME

print("\n --- Simulation Summary ---")
print(f"Total Vehicles Entered: {total_vehicles}")
print(f"Vehicles Exited: {exited_vehicles}")
print(f"Average Congestion: {congestion*100:.2f}%")
print(f"Average Wait Time: {avg_wait_time:.2f} steps")
print(f"Traffic Flow: {flow:.2f} vehicles/time step")
```

# Ques 1:  
WAP that creates a list of 10 random integers. Then create two list odd list and even list that has all odd and even values in .

```python
import random
numbers = [random.randint(1, 100) for _ in range(10)]
def is_even(x):
    return x % 2 == 0
def is_odd(x):
    return x % 2 != 0
even_list = list(filter(is_even, numbers))
odd_list = list(filter(is_odd, numbers))
print("Original list:", numbers)
print("Even numbers:", even_list)
print("Odd numbers:", odd_list)
```

# Ques 2:  
WAP a program that converts strings of all uppercase characters into strings of all lowerscase characters using map().

```python
def lower(x):
    return x.lower()
words = ["HELLO", "PYTHON", "PROGRAMMING","Artificial Intelligence"]
lowercase_words = list(map(lower, words))
print("Original list:", words)
print("Lowercase list:", lowercase_words)
```

# Ques 3:  
WAP to find the largest value in a list using reduce().

```python
from functools import reduce
numbers = [12, 45, 67, 23, 89, 34, 90, 11]
largest = reduce(lambda x, y: x if x > y else y, numbers)
print("List:", numbers)
print("Largest value:", largest)
```

# Ques 6:  
WAP to find intersection,union,symmetric difference b/w two sets: coders and analysts .
1) people working as coders as well as analysts.
2) People working as coders or analysts.
3) People working as analysts but not coders.
4) People working in only one of the groups.

```python

coders = {"Atul", "Neha", "Ravi", "Priya", "Ankit"}
analysts = {"Ravi", "Ankit", "Sonia", "Meena"}
both = coders & analysts
print("1) Coders as well as Analysts:", both)

either = coders | analysts
print("2) Coders or Analysts:", either)


analysts_only = analysts - coders
print("3) Analysts but not Coders:", analysts_only)


only_one = coders ^ analysts
print("4) Only one of the groups:", only_one)
```
