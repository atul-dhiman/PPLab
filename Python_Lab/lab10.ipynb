1. Write a Python program to create and display all combinations of letters, selecting each
letter from a different key in a dictionary.
Sample data: {&#39;1&#39;:[&#39;a&#39;,&#39;b&#39;], &#39;2&#39;:[&#39;c&#39;,&#39;d&#39;]}
Expected Output:
ac
ad
bc

```python
import itertools

data = {'1': ['a','b'], '2': ['c','d']}

# Get combinations
keys = list(data.keys())
values = [data[k] for k in keys]

for combo in itertools.product(*values):
    print("".join(combo))
```

2. Write a Python program to get the top three items in a shop.
Sample data: {&#39;item1&#39;: 45.50, &#39;item2&#39;:35, &#39;item3&#39;: 41.30, &#39;item4&#39;:55, &#39;item5&#39;: 24}
Expected Output:
item4 55
item1 45.5
item3 41.3

```python
shop = {'item1':45.50, 'item2':35, 'item3':41.30, 'item4':55, 'item5':24}

# Sort dictionary by values (descending)
top_three = sorted(shop.items(), key=lambda x: x[1], reverse=True)[:3]

for item, price in top_three:
    print(item, price)
```

3. Write a Python program to find the specified number of maximum values in a given
dictionary.
Original Dictionary:
{&#39;a&#39;: 5, &#39;b&#39;: 14, &#39;c&#39;: 32, &#39;d&#39;: 35, &#39;e&#39;: 24, &#39;f&#39;: 100, &#39;g&#39;: 57, &#39;h&#39;: 8, &#39;i&#39;: 100}
1 maximum value(s) in the said dictionary:
[&#39;f&#39;]
2 maximum value(s) in the said dictionary:
[&#39;f&#39;, &#39;i&#39;]
5 maximum value(s) in the said dictionary:
[&#39;f&#39;, &#39;i&#39;, &#39;g&#39;, &#39;d&#39;, &#39;c&#39;]

```python
data = {'a':5, 'b':14, 'c':32, 'd':35, 'e':24, 'f':100, 'g':57, 'h':8, 'i':100}

def n_max_values(d, n):
    # Sort by values in descending order
    sorted_items = sorted(d.items(), key=lambda x: x[1], reverse=True)
    return [k for k, v in sorted_items[:n]]

print("1 maximum value:", n_max_values(data, 1))
print("2 maximum values:", n_max_values(data, 2))
print("5 maximum values:", n_max_values(data, 5))
```

4. Write a Python program to combine two lists into a dictionary. The elements of the first one
serve as keys and the elements of the second one serve as values. Each item in the first list
must be unique and hashable.
Original lists:
[&#39;a&#39;, &#39;b&#39;, &#39;c&#39;, &#39;d&#39;, &#39;e&#39;, &#39;f&#39;]
[1, 2, 3, 4, 5]
Combine the values of the said two lists into a dictionary:
{&#39;a&#39;: 1, &#39;b&#39;: 2, &#39;c&#39;: 3, &#39;d&#39;: 4, &#39;e&#39;: 5}

```python
keys = ['a','b','c','d','e','f']
values = [1, 2, 3, 4, 5]

result = dict(zip(keys, values))

print(result)
```

5. Write a Python function that evaluates mathematical expressions represented as
dictionaries. The expression dictionary should support basic operations (+, -, *, /), variables,
and nested expressions.

```python
{
    "op": "+",
    "left": 5,
    "right": {
        "op": "*",
        "left": "x",
        "right": 3
    }
}
```

```python
def evaluate(expr, variables=None):
    if variables is None:
        variables = {}

    # If expr is a number
    if isinstance(expr, (int, float)):
        return expr
    
    # If expr is a variable
    if isinstance(expr, str):
        return variables[expr]
    
    # Otherwise it is a dictionary expression
    op = expr["op"]
    left = evaluate(expr["left"], variables)
    right = evaluate(expr["right"], variables)

    if op == "+":
        return left + right
    elif op == "-":
        return left - right
    elif op == "*":
        return left * right
    elif op == "/":
        return left / right
    else:
        raise ValueError("Unknown operator")

# Example expression
expr = {
    "op": "+",
    "left": 10,
    "right": {
        "op": "*",
        "left": "x",
        "right": 5
    }
}

# Evaluate with variable x = 2
print(evaluate(expr, {"x": 2}))
```
