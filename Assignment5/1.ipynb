Q1. Write a program that has classes such as Student, Course and Department.
Enroll a student in a course of a particular department.

```python
class Department:
    def __init__(self, name):
        self.name = name
        self.courses = []

    def add_course(self, course):
        self.courses.append(course)


class Course:
    def __init__(self, course_name, dept):
        self.course_name = course_name
        self.department = dept
        self.students = []
        dept.add_course(self)

    def enroll_student(self, student):
        self.students.append(student)
        print(f"{student.name} enrolled in {self.course_name} ({self.department.name})")


class Student:
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll


# ---- Demo ----
d1 = Department("Computer Science")
c1 = Course("Python Programming", d1)
s1 = Student("Atul", 101)

c1.enroll_student(s1)
```

Q2. Write a program with class Bill. The users have the option to pay the bill
either by cheque or by cash. Use the inheritance to model this situation.

```python
class Bill:
    def __init__(self, amount):
        self.amount = amount

    def pay(self):
        pass


class CashPayment(Bill):
    def pay(self):
        print(f"Paying ₹{self.amount} by Cash.")


class ChequePayment(Bill):
    def __init__(self, amount, cheque_no):
        super().__init__(amount)
        self.cheque_no = cheque_no

    def pay(self):
        print(f"Paying ₹{self.amount} by Cheque No: {self.cheque_no}")


# ---- Demo ----
b1 = CashPayment(5000)
b1.pay()

b2 = ChequePayment(12000, "CHQ12345")
b2.pay()
```

Q3.Write a program to overload the - =operator to subtract two Distance
Objects.

```python
class Distance:
    def __init__(self, meters):
        self.meters = meters

    def __isub__(self, other):
        self.meters = self.meters - other.meters
        return self

    def display(self):
        print("Distance =", self.meters, "meters")


# ---- Demo ----
d1 = Distance(120)
d2 = Distance(50)

d1 -= d2
d1.display()     # Output: Distance = 70 meters
```

Q4. Write a program with two classes for calculating the distance. On has
distance specified in meters and other has distance in kilometres. These
functions take argument of class Distance that converts the distance into
kilometres and meters.

```python
class MeterDistance:
    def __init__(self, meters):
        self.meters = meters

    def to_km(self):
        return self.meters / 1000


class KilometerDistance:
    def __init__(self, km):
        self.km = km

    def to_meters(self):
        return self.km * 1000


# ---- Function to display conversions ----
def convert_distance(obj):
    if isinstance(obj, MeterDistance):
        print(f"{obj.meters} meters = {obj.to_km()} km")
    elif isinstance(obj, KilometerDistance):
        print(f"{obj.km} km = {obj.to_meters()} meters")


# ---- Demo ----
m = MeterDistance(1500)
k = KilometerDistance(3.2)

convert_distance(m)
convert_distance(k)
```
